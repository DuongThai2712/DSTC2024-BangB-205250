Thống kê 5 mặt hàng có tổng thời gian nhìn và cầm xem lâu nhất?

```python
item_time = df_customer.groupby(['Shelf ID','Item ID']).agg(
    total_look=('Looking at item (s)', 'sum'), 
    total_hold=('Holding the item (s)', 'sum')
)
item_time['total_interaction'] = item_time['total_look'] + item_time['total_hold']

item_time.sort_values('total_interaction', ascending=False).merge(df_item_info[['Shelf ID', 'Item ID','Name']], on=['Shelf ID','Item ID']).head(5)
```

Thống kê 5 mặt hàng thường được cầm lên rồi trả lại nhiều nhất?

```python
item_return = df_customer[(df_customer['Picking up item']==True) & (df_customer['Returning item']==True)].groupby(['Shelf ID','Item ID']).agg(
    total_pickup=('Picking up item', 'count'), 
    total_return=('Returning item', 'count')
)
item_return = item_return.sort_values('total_return', ascending=False).merge(df_item_info[['Shelf ID','Item ID','Name']], on=['Shelf ID','Item ID'])
```

```python
item_return.head(5)
```

Các nhóm khách hàng theo độ tuổi (Thiếu niên: 18 - 30; Trung niên: 31 - 60; Cao tuổi: > 60) mua mặt hàng nào nhiều nhất?

```python
# Lọc thông tin khách hàng không rõ ràng về việc cho vào giỏ hay lấy ra (thiếu dữ liệu)
df_customer_bought = df_customer[(df_customer['Putting item into bag'].isnull()==False) &
                                 (df_customer['Putting item into bag in the 2nd time'].isnull()==False) &
                                 (df_customer['Taking item out of bag'].isnull()==False)
                                ]
```

```python
df_customer_bought = df_customer_bought[((df_customer_bought['Taking item out of bag']==True) & (df_customer_bought['Putting item into bag in the 2nd time']==True)) | ((df_customer_bought['Taking item out of bag']==False) & (df_customer_bought['Putting item into bag']==True))]
```

```python
df_customer_bought.head()
```

```python
df_customer_teen = df_customer_bought[(df_customer['Age'] >= 18) & (df_customer['Age'] <= 30)]
df_customer_mid = df_customer_bought[(df_customer['Age'] >= 31) & (df_customer['Age'] <= 60)]
df_customer_old = df_customer_bought[df_customer['Age'] > 60]
```

```python
item_bought_teen = df_customer_teen.groupby(['Shelf ID','Item ID']).agg(total=('Item ID', 'count'))
item_bought_teen.sort_values('total', ascending=False).merge(df_item_info[['Shelf ID','Item ID', 'Name']], on=['Shelf ID','Item ID']).head(1)
```

```python
item_bought_mid = df_customer_mid.groupby(['Shelf ID','Item ID']).agg(total=('Item ID', 'count'))
item_bought_mid.sort_values('total', ascending=False).merge(df_item_info[['Shelf ID','Item ID', 'Name']], on=['Shelf ID','Item ID']).head(1)
```

```python
item_bought_old = df_customer_old.groupby(['Shelf ID','Item ID']).agg(total=('Item ID', 'count'))
item_bought_old.sort_values('total', ascending=False).merge(df_item_info[['Shelf ID','Item ID', 'Name']], on=['Shelf ID','Item ID']).head(1)
```

Top 3 quầy hàng có thời lượng trung bình quan tâm đến sản phẩm, trên số lượt tương tác, là lâu nhất (quan tâm tương ứng với việc nhìn và cầm xem)?

```python
df_customer_cares = df_customer[['Shelf ID', 'Looking at item (s)', 'Holding the item (s)']]
df_customer_cares.head()
```

```python
df_customer_cares['cares'] = df_customer_cares['Holding the item (s)'] + df_customer_cares['Looking at item (s)']
df_customer_cares.head()
```

```python
df_customer_cares.drop(['Looking at item (s)', 'Holding the item (s)'], axis=1)
```

```python
avg_shelf = df_customer_cares.groupby('Shelf ID').agg(
    counts=('Shelf ID', 'count'), 
    cares=('cares', 'sum')
)
```

```python
avg_shelf['avg']= avg_shelf['cares'] / avg_shelf['counts']
```

```python
avg_shelf.merge(df_shelf_info[['Shelf ID', 'Description']], on='Shelf ID')
```
